```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt

data = pd.read_csv('output.csv', parse_dates=["Date"], index_col=None)

posCloud = pd.read_csv('posCloud.csv')
negCloud = pd.read_csv('negCloud.csv')

sentimentSeries = data.plot(y=0, x=2, figsize=(15,10))
sentimentSeries.get_figure().savefig("images/sentiment.png")
```

```python
sentimentSeries = data.plot(y=1, x=2, figsize=(15,10))
sentimentSeries.get_figure().savefig("images/ratio.png")
```

```python
new = pd.read_csv('output.csv', parse_dates=["Date"], index_col=None)
new.set_index('Date', inplace=True)
score_opinion = new.groupby('Opinion')['Sentiment'].plot(legend=True, figsize=(15,10))
plt.ylabel("Sentiment Score")
plt.savefig("images/score_opinion.png")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

posWordFreq = dict(zip(posCloud["Word"].tolist(), posCloud["Frequency"].tolist()))

# print(wordFreq)

# Manually removed the word 'sincerely' as a stop word since we cannot use stopwords with generate_from_fr
posWordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(posWordFreq)

plt.figure(figsize=(10,10))
plt.imshow(posWordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

posWordcloud.to_file("images/posCloud.png")

```

```python
negWordFreq = dict(zip(negCloud["Word"].tolist(), negCloud["Frequency"].tolist()))

# print(wordFreq)
negWordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(negWordFreq)

plt.figure(figsize=(10,10))
plt.imshow(negWordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

negWordcloud.to_file("images/negCloud.png")
```
